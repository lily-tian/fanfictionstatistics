```python
# imports libraries
import pickle						# import/export lists
import re 							# regular expression
import pandas as pd					# dataframes
import numpy as np					# numerical computation
import matplotlib.pyplot as plt		# plot graphics
```

```python
# opens raw data
with open ('df_profile', 'rb') as fp:
    df_profile = pickle.load(fp)
```

```python
# examines when users first joined
entry_year = df_profile.loc[df_profile.join_year != 'NaN', 'join_year']
entry = entry_year.value_counts().sort_index()
entry.plot()

plt.show()
```

```python
# examines status of users
status = df_profile['status'].value_counts()
status.plot.bar()
plt.xticks(rotation=0)

plt.show()
```

```python
# examines distribution of top 10 countries
country = df_profile['country'].value_counts()[1:10]
country.plot.bar()
plt.xticks(rotation=45)

plt.show()
```

```python
# examines distribution of stories written
st = pd.to_numeric(df_profile.loc[df_profile.status == 'author', '#st'])
st.plot.hist()

plt.show()
```

```python
# examines distribution of favorited
fav = df_profile.loc[df_profile.status != 'inactive', ['#fs', '#fa']].apply(pd.to_numeric)
fav = fav.loc[fav['#fs'] > 0 | fav['#fa'] > 0, ]
plt.figure()
fav.plot.hist(alpha=0.5, range=[0, 10])

plt.show()
```
